```
import requests
import pytz
import json
import os
from datetime import datetime,timezone, timedelta
from zipfile import ZipFile
import time
```

```
"""
DEMONIO DATOS ESPIRAS
Este demonio descarga los datos de tramos, dónde se encuentra el tramo que
nos interesa (Paso inferior V-21, salida BARCELONA)
"""

#Esta función descarga el objeto json y lo guarda en un format
#Se ejectua cada quince minutos haciendo uso de CRONTAB (UNIX)
def downloadJSON():
  url = "http://apigobiernoabiertortod.valencia.es/apirtod/datos/intensidad_tramos.json?items=1000"
  response = requests.get(url)
  if response.status_code == 200:
    tz = pytz.timezone('Europe/Madrid')
    time = datetime.now(tz)
    current_time = time.strftime("%d-%m-%Y_%H-%M-%S")
    current_time = str(current_time)
    filename = './espiras/espiras_'+current_time+'.json'
    with open(filename,mode='w',encoding='UTF-8') as f:
      json.dump(response.json(),f)
  return


if __name__ == '__main__':
  downloadJSON()
#-----------------------------------------------------------------------
#Estas dos funciones se ejecutan cada 1 hora

#Esta función se encarga de comprimir los archivos y guardarlos
def zipEspiras():
  dir = './espiras/'
  fileNames = os.listdir(dir)
  with ZipFile('espiras.zip', 'a') as f:
    for fileName in fileNames:
      filePath = './espiras/'+fileName
      f.write(filePath)
  return

#Esta función se encarga de eliminar los archivos comprimidos
def delEspiras():
  dir = './espiras/'
  filenames = os.listdir(dir)
  for archivo in filenames:
    archivo = "./espiras/"+archivo
    os.remove(archivo)
  return

if __name__ == '__main__':
  zipEspiras()
  time.sleep(5)
  delEspiras()
```
